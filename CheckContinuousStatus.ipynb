```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime 
```

```python
sleepDf = pd.read_csv("/Users/hyennaeon/Desktop/Hatinroo/HatinrooData/수면_활동_부재_수집데이터/SleepData.csv")
absenceDf = pd.read_csv("/Users/hyennaeon/Desktop/Hatinroo/HatinrooData/수면_활동_부재_수집데이터/AbsenceData.csv")
activeDf = pd.read_csv("/Users/hyennaeon/Desktop/Hatinroo/HatinrooData/수면_활동_부재_수집데이터/ActiveData.csv")
```

```python
def checkCount(df, value1, value2):
    count = 0
    consecutive_count = 0

    for row in df['status']:
        if row == value1 and row == value2:
            count += 1
            if count >= 180:
                consecutive_count += 1
        else:
            count = 0

    return consecutive_count

df = sleepDf[sleepDf['userId'] == 1]
consecutive_count = checkCount(df, 3, 4)
print("횟수:", consecutive_count)
```

```python
status = "부재"
nameArray = ['하원','효은','민우','명균','지원','정민','채별']
for i in range(7):
    Df = activeDf[activeDf['userId'] == i]
    ratio = [len(Df[Df['status'] != 0]), len(Df[Df['status'] == 0])]
    labels = ['신뢰', '비신뢰']
    plt.title(nameArray[i] + " " + status)
    plt.pie(ratio, labels=labels, autopct='%.1f%%')
    plt.savefig('/Users/hyennaeon/Desktop/Hatinroo/HatinrooData/'+nameArray[i]+".png")
    plt.show()
```
